#### Compare best years 2017 - 2018 to the last 2 years (2022 - 2023)
2017 - 2018 is when the market has had its highest level of production. Let's investigate which difference there are between now and then.

```python
# Filter for the years 2017 and 2018
prod_df_filtered_17_18 = production_data_df[(production_data_df['ProducingMonth'].dt.year == 2017) | (production_data_df['ProducingMonth'].dt.year == 2018)]

# Filter for the years 2022 and 2023
prod_df_filtered_22_23 = production_data_df[(production_data_df['ProducingMonth'].dt.year == 2022) | (production_data_df['ProducingMonth'].dt.year == 2023)]

# Plot 2017 and 2018
plt.figure(figsize = (14, 6))
sns.lineplot(data = prod_df_filtered_17_18,
             x='ProducingMonth',
             y='Prod_BOE')

# Plot 2022 and 2023
plt.figure(figsize = (14, 6))
sns.lineplot(data = prod_df_filtered_22_23,
             x='ProducingMonth',
             y='Prod_BOE')
```

Comparing the 2-year spans, similar pattern are spotted in the first year for both cases. Although bening volatile, both 2017 and 2022 have had higher levels of barrels produced compared to the year after. The main difference is that in 2018 there has been a slightly production increase, while in 2023 the trend seems to keep dropping.

##### Monthly seasonality
Comparing the best years to the last 2 years to see if, despite the total level of production, seasonal trends are the same or changed (before and after Covid).

```python
# Combine both datasets into a single DataFrame
combined_df = pd.concat([prod_df_filtered_17_18, prod_df_filtered_22_23])

# Extract month information
combined_df['Month'] = combined_df['ProducingMonth'].dt.month_name()

# Plot both datasets on the same graph
plt.figure(figsize=(14, 6))

# Use sns.lineplot for 2017-2018
sns.lineplot(data=combined_df[combined_df['ProducingMonth'].dt.year.isin([2017, 2018])],
             x='Month',
             y='Prod_BOE',
             label='2017-2018')

# Use sns.lineplot for 2022-2023
sns.lineplot(data=combined_df[combined_df['ProducingMonth'].dt.year.isin([2022, 2023])],
             x='Month',
             y='Prod_BOE',
             label='2022-2023')

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Prod_BOE')
plt.title('Comparison of Prod_BOE for 2017-2018 and 2022-2023')

# Show legend
plt.legend()
```

The seasonal patterns remained the same. Although less remarkable, peaks of production in October are still present. The summer (Jun - Aug) have a low and steady increase, folloed by a slightly drop in September.
Between January and March the biggest decrease in production is recorded.

##### Best months of production for Trading purposes:
1. May, August, October, December.
##### Worst peiods of production for Trading purposes:
2. January, February, March, June.

2. Financial Stability: Selling drilling rights to financially robust companies reduces the risk of payment defaults or delays in receiving revenues from the sale of rights.
3. Operational Capability: Established operators with a strong track record usually possess advanced technological capabilities, skilled personnel, and efficient operational processes (successful and efficien dirlling operations)
4. Risk Mitigation: Bigger operators often have diversified portfolios and can better absorb risks associated with exploration and production activities. This can be beneficial for the trading company as it reduces exposure to potential operational and market risks.
